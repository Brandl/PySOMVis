```python
from fuzzywuzzy import fuzz
import pandas as pd
df = pd.read_excel('C:\\Users\\admin\\Desktop\\НОМЕНКЛУТАР\\ФИЛИАЛ\\НоменклатураФилиал.xlsx')
combined_list = df.iloc[:,6].tolist()
grs = list()
number = len(combined_list)
while number>0:
    newlist = list()
    name = combined_list[0]
    for i in range(len(combined_list)):
        if fuzz.token_sort_ratio(name, combined_list[i]) > 60:
            newlist.append(combined_list[i])
    grs.append(newlist)
    for cc in newlist:
        while cc in combined_list:
            combined_list.remove(cc)
    number = len(combined_list)
```

```python
index = 0;
number = 0
l = []
total = 0
for i in range(len(grs)):
    l.append(len(grs[i]))
```

```python
duplicate_dict={} # a dictionary to store each of them.
for i in l:#loop through them.
    duplicate_dict[i]=l.count(i)
print(duplicate_dict)
```

```python
sorted(duplicate_dict.items(), key=lambda x: x[1], reverse=True)
```

```python
import xlsxwriter
from functools import reduce

writer = pd.ExcelWriter('D:\\Dropbox\!Master thesis\\implementations\\SOMFinalizationLib\\SOMToolbox\\maps\\goups.xlsx', engine='xlsxwriter')
group = 0
for dub in duplicate_dict:
    n = []
    for g in grs:
        if len(g)==dub:
            n.append(g)
    df = pd.DataFrame(reduce(lambda x, y: x+y, n))
    df.to_excel(writer, sheet_name='group_'+str(len(dub)), index=False)
    #group += 1
writer.save()
```
